```python
df_filtered = df[df['CustomerID'] != 0.0].copy()
df_filtered['TotalSpend'] = df_filtered['Quantity'] * df_filtered['UnitPrice']

customer_segmentation_df = df_filtered.groupby('CustomerID').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalSpend=('TotalSpend', 'sum')
).reset_index()

print(customer_segmentation_df.head())
print(customer_segmentation_df.info())
```

```python
from sklearn.preprocessing import StandardScaler
print("StandardScaler imported successfully.")
```

```python
X = customer_segmentation_df[['TotalQuantity', 'TotalSpend']]
print("DataFrame X created with 'TotalQuantity' and 'TotalSpend' columns.")
```

```python
scaler = StandardScaler()
print("StandardScaler object instantiated.")
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
print("Features scaled and stored in X_scaled_df.")
print(X_scaled_df.head())
```

```python
from sklearn.cluster import KMeans
print("KMeans imported successfully.")
```

```python
wcss = []
k_values = range(1, 11) # Define a range of K values from 1 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X_scaled_df)
    wcss.append(kmeans.inertia_)

print("WCSS values calculated successfully for K from 1 to 10.")
print(wcss)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid(True)
plt.show()
```

```python

optimal_k = 3

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
customer_segmentation_df['Cluster'] = kmeans.fit_predict(X_scaled_df)

print(customer_segmentation_df.head())
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='TotalQuantity', y='TotalSpend', hue='Cluster', data=customer_segmentation_df, palette='viridis', s=100, alpha=0.8)
plt.title(f'Customer Segments (K={optimal_k})')
plt.xlabel('Total Quantity')
plt.ylabel('Total Spend')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X_scaled_df, customer_segmentation_df['Cluster'])

print(f"The Silhouette Score for the clustering is: {silhouette_avg:.3f}")
```
